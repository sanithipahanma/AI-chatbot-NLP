**Data Preprocessing function**

```python
def preprocess_text (user_input):

  #lowercase
  text = user_input.lower()

  #Punctuation removal
  text = re.sub(r'[^a-z0-9\s]', '', text)

  #Extra space removing
  text = re.sub(r'\s+', ' ', text).strip()

  #spelling correction
  text_corrected = str(TextBlob(text).correct())

  #remove stopwords
  stop_words = set(stopwords.words('english'))
  text_no_stop = ' '.join([w for w in text.split() if w not in stop_words])

  #Tokenization
  tokens = word_tokenize(text_no_stop)
  return tokens


```

**Main function of the whole model**

```python
def chatbot (my_chatbot, knowledgebase):
  print ("---------------------------------------------------------------------------------------------------------------")
  print ("                                       Hello! Welcome to "business name"!")
  print ("---------------------------------------------------------------------------------------------------------------\n")
  print ("Berry: Type 'bye' to exit or 'help' to view Menu options.\n")


  file_name = "/content/drive/MyDrive/DonBerry_Chat_History.csv"

  file_exist = os.path.isfile(file_name)

  with open (file_name, "a") as file:
    writer = csv.writer(file)

    if not file_exist:
      writer.writerow (["Timestamp", "User_Input", "Berry_Response"])

    menu = """Berry: You can ask me about: \n
    - Store Hours
    - Location or Address
    - Products or services
    - Payment Methods
    - Delivery or Delivery charge
    - Customize products
    - Exchange policies
    - 'bye' to exit
    """

    while True:
      user_input = input("You: ")
      ending = ["bye", 'exit', "quit"]
      response1 = "Goodbye! Have a nice day!"
      if any(user_input.lower().startswith(x) for x in ending):
        print (f"Berry: {response1}")
        writer.writerow([
          datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
          user_input, response1])
        break

      if user_input.lower() in ["help", "menu", "options"]:
        print (menu)
        writer.writerow([
          datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
          user_input, menu])
        continue

      processed = preprocess_text(user_input)
      token_set = set(processed)

      found = False

      for keywords, response in knowledgebase:
        for phrase in keywords:
          if phrase in token_set:
            print ("Berry: ", response)
            found = True
            break
        if found:
          break

      if not found:
        response = my_chatbot.respond (user_input)
        if response:
          print ("Berry: ", response)

        else:
          response = """Sorry, I couldn't Understand that.
        Are you asking about products, location, delivery or payments.
        Tip: Type 'help' to view the menu.\n"""
          print ("Berry: ", response)

      writer.writerow([
          datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
          user_input, response
      ])
```
